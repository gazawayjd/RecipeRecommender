# Recipe Recommendation

### Enter your pantry ingredients separated by commas (,) to generate recipes that best fit your pantry!

```python
import pandas as pd
import numpy as np
import nltk
import string
import ast
import re
import unidecode
import spacy
import ipywidgets as widgets
from IPython.display import display, clear_output
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Create dataframe from recipe database
df = pd.read_csv('./recipes_parsed1.csv', index_col='RecipeID')
df['RecipeInstructions'] = df['RecipeInstructions'].str.replace('\n', '')

# Function to clean user inputs
def ingredient_cleanup(ingredients):
    terms_to_remove = ['paso','el','great','salt', 'pepper', 'oil','of', 'zest','fresh', 'butter', 'sugar', 'water', 'cloves', 'all','purpose', 'black', 'pepper', 'garlic', 'powder', 'cinnamon', 'juice', 'paprika','clove', 'chili','ground','extra','virgin','granulated','cumin','dried','oregano', 'kosher', 'boneless','skinless','canola', 'basil', 'white','thyme','cilantro','vinegar','powder','confectioners','seasoning','lean','curry','light','crushed','dry','boiling','sea', 'plain']
    
    # initialize lemmatizer and ingredient list to be returned
    lemmatizer = WordNetLemmatizer()
    cleaned_ingredients = []

    # Loop to iterate through each word in user input
    for i in ingredients:
        if not isinstance(i, str):
            i = str(i)

        # Split words according to spaces and hyphens
        ingr = re.split(' |-', i)

        # Remove any non-alphabetic terms, set it lower-case and remove any accents
        ingr = [word for word in ingr if word.isalpha()]
        ingr = [word.lower() for word in ingr]
        ingr = [unidecode.unidecode(word) for word in ingr]

        # Lemmatize each word for analysis and comparison
        ingr =[lemmatizer.lemmatize(word) for word in ingr]
        # remove common term ingredients listed in terms_to_remove
        ingr = [word for word in ingr if word not in terms_to_remove]

        if ingr:
            cleaned_ingredients.append(' '.join(ingr))
    cleaned_ingredients_str = ' '.join(cleaned_ingredients)
    return [cleaned_ingredients_str]

# Generate recommendations
def get_cosine_sim_score(user_input, df):
    # Create a CountVectorizer to convert text data into a numerical format
    tfidf = TfidfVectorizer()

    # Fit and transform the user input and recipe ingredients
    user_input_vector = tfidf.fit_transform(user_input)
    recipe_vectors = tfidf.transform(df['RecipeIngredientParts'].values.astype('U'))

    # Calculate cosine similarity between user input and recipe ingredients
    similarities = cosine_similarity(user_input_vector, recipe_vectors)
    similarity_scores = similarities[0]

    # Create a DataFrame to store the similarity scores
    similarity_df = pd.DataFrame({'Cosine Similarity:': similarity_scores})

    # Add a column for recipe names or IDs (assuming you have one)
    similarity_df['Name'] = df['Name']

    # Sort dataframe by highest score
    highest_similarity_df = similarity_df.sort_values(by='Cosine Similarity:', ascending=False)
    
    return highest_similarity_df.head(5)
    

# Function to handle user click and show top 5 recipes per similarity score
def recommend_recipes(b):
    with output:
        output.clear_output()  # Clear previous results
        ingredients = ingredient_input.value.strip().split(',')    
        
        global top_five_df
        try:
            top_five_df = get_cosine_sim_score(ingredient_cleanup(ingredients), df)
        except:
            print("Invalid input, please try again")
        
        print('Your Pantry: ' + str(ingredients) + '\n')
        print(get_cosine_sim_score(ingredient_cleanup(ingredients), df))
```

```python
# Create a label for Pantry title
pantry_label = widgets.Label(value="Pantry Inventory:", style=dict(font_style='bold', font_size='24px', text_color='Red', text_decoration='underline'))

# Create a text area widget for users to enter pantry ingredients
ingredient_input = widgets.Textarea(
    value='',
    placeholder='Enter ingredients separated by commas (e.g. eggs, milk, flour)',
    description='',
    layout=widgets.Layout(width='500px', height='100px')
)

# Create a button for submitting the ingredients
submit_button = widgets.Button(description='Submit Ingredients')

# Create an output widget to display recommendations
output = widgets.Output(layout = widgets.Layout(width='750px'))

# Controller for button click
submit_button.on_click(recommend_recipes)

# Create an HBox to align the text area and button horizontally
input_box = widgets.VBox([pantry_label,ingredient_input, submit_button, output])

# Display the widgets
display(input_box)
```

## See below for your recommended recipe instructions!

```python
def get_top_five_instructions (top_five_df):
    # Create list for recipe names, ingredients, and instructions to append to
    instr_list = []
    
    for i in top_five_df.index:
        instructions = str(df['RecipeInstructions'].iloc[i])
        instructions_cleaned = instructions.replace('", ', '\n').replace('"','')
        recipe_name = str(df['Name'].iloc[i])
        recipe_ingreds = str(df['RecipeIngredientParts'].iloc[i])
        
        instr_list.append(''.join(recipe_name))
        instr_list.append(''.join('Ingredients: ' + recipe_ingreds))
        instr_list.append(''.join(instructions_cleaned + '\n'))
    
    display_instructions = "\n".join(instr_list)
    return display_instructions

# Function to handle button click and display instructions for top 5 recipes
def show_instructions(b):
    with output1:
        clear_output()  # Clear previous instructions if any
        instructions_display.value = get_top_five_instructions(top_five_df)

# Create text area to display instructions
instructions_display = widgets.Textarea(
    value='',
    placeholder='',
    description='',
    layout=widgets.Layout(width='750px', height='500px')
)            

# Create a button for showing instructions
show_button = widgets.Button(description='Show Instructions')

# Controller for button click
show_button.on_click(show_instructions)

# Create an output widget to display instructions
output1 = widgets.Output(layout = widgets.Layout(width='750px'))

input_box = widgets.VBox([show_button,instructions_display])

display(input_box)
display(output1)
```

## Histogram Of Top 20 Ingredients In Recipe Database

```python
# Histogram plotting 20 most common ingredients in recipe database
import seaborn as sns
import matplotlib.pyplot as plt

vocab = nltk.FreqDist()
for ingredients in df['RecipeIngredientParts'].astype(str):
    ingredients = ingredients.split()
    vocab.update(ingredients)

top20ingred_fdist = vocab.most_common(20)
top20ingred_fdist = pd.Series(dict(top20ingred_fdist))
figure, ax = plt.subplots(figsize=(10,10))

all_plot = sns.barplot(x=top20ingred_fdist.index, y=top20ingred_fdist.values, ax=ax)
plt.xticks(rotation = 30)
plt.show()
```

## Scatterplot showing number of ingredients per recipe compared to Review Count

```python
# Count the number of ingredients for each recipe
df['ingredient_count'] = df['RecipeIngredientParts'].astype(str).apply(lambda x: len(x.split()))

sns.scatterplot(x=df['ingredient_count'], y=df['ReviewCount'], data=df)
plt.show()
```

## Scatterplot showing most common recipe categories

```python
import seaborn as sns
import matplotlib.pyplot as plt
vocab = nltk.FreqDist()
for categories in df['RecipeCategory'].astype(str):
    categories = categories.split('\n')
    vocab.update(categories)

top20categories_fdist = vocab.most_common(20)
top20categories_fdist = pd.Series(dict(top20categories_fdist))
figure, ax = plt.subplots(figsize=(10,10))

all_plot = sns.barplot(x=top20categories_fdist.index, y=top20ingred_fdist.values, ax=ax)
plt.xticks(rotation = 30)
plt.show()
```
